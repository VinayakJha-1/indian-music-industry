## Song Data Analysis

```python
import pandas as pd 
import numpy as np
import matplotlib as mtb
import seaborn as sns
```

```python
# Read the data 
df = pd.read_csv('../data/dummy_song_data.csv')
```

```python
df.shape
```

```python
df
```

```python
# Add date and year 
df['release_date'] = pd.to_datetime(df['release_date'], dayfirst=True)
df
```

```python
# Add  decade year 
df['year'] = df['release_date'].dt.year
df['decade'] = df['year'] - df['year']%10
```

```python
## Checking popularity index groupby decade mean
x = df.groupby('decade')['popularity'].mean()
x
```
