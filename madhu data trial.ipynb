```python
stripped = df.drop(["name", "album", "artist", "year", "release_date", "mode"], axis=1)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=7, random_state=0).fit(stripped)
kmeans.labels_
kmeans.cluster_centers_
y_kmeans = kmeans.predict(stripped)
```

```python
from collections import defaultdict
s = defaultdict(list)

for i in range(len(y_kmeans)):
    s[y_kmeans[i]].append(df["name"][i])
    
for x in s.values():
    print(x)
    print(1)
```
